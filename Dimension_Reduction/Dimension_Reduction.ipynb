## Umap

```python
import umap
from sklearn.datasets import load_digits

digits = load_digits()

embedding = umap.UMAP(n_neighbors=5,
                      min_dist=0.3,
                      metric='correlation').fit_transform(digits.data)
```

```python
digits
```

```python
import pandas as pd

df_umap = pd.DataFrame(digits.data).copy()
df_umap['label'] = digits.target.copy()
df_umap['label_str'] = df_umap['label'].apply(lambda x: str(x))
```

```python
df_umap
```

```python
from ggplot import *

df_umap['x-umap'] = embedding[:,0]
df_umap['y-umap'] = embedding[:,1]


chart = ggplot( df_umap, aes(x='x-umap', y='y-umap', color='label_str') ) \
        + geom_point(size=70,alpha=0.5) \
        + ggtitle("umap dimensions colored by digit")
chart
```
